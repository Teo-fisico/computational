```python
import numpy as np
import matplotlib.pyplot as plt

# Aqui escreva uma nova função
def tfun(x):
    return x**4-x**2+x/4

#escreva sua derivada da função anterior

def der_tfun(x):
    return 4*x**3-2*x+0.25

```

```python
# Parâmetros de gradiente
x=-2.03
taxa_ap=0.1
toler=0.01
nu_itera=150
```

```python
#salvar os dados
x_sal=[x]
for i in range(nu_itera):
    grad=der_tfun(x)
    x_no=x-taxa_ap*grad

    x_sal.append(x)
    if abs(x_no-x)<toler:
        print(f"Converge em iteração:{i+1}")
        break

    x=x_no
```

```python
# mostar valore
min_x=x
min_fx=tfun(min_x)

#print(f"x final:{min_x}")
print(f"f(x) aproximado:f({(min_x)})={tfun(min_x)}")
```

```python
# Plotar a função
x_valor=np.linspace(-2.1,2.1,200)
y_valor=tfun(x_valor)

plt.figure(figsize=(8,6))
plt.plot(x_valor,y_valor,label='f(x)',color='blue')

# marcar os passos de GD
x_sa=np.array(x_sal)
y_sa=tfun(x_sa)

plt.plot(x_sa,y_sa,'ro--',label='Gradiente descendente')

# anotar o ponto final
plt.scatter([x_sa[-1]],[y_sa[-1]],color='green',s=100,label='Último Ponto')
plt.title('Questão 3: Gradiente descendente f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
```
