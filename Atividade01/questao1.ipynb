```python
import numpy as np
import matplotlib.pyplot as plt

# Aqui defina sua função preferida
def tb(x):
    return x**2-1

# Escreva sua derivada da função anterior
def der_tb(x):
    return 2*x  
```

```python
# defina os parâmetros de cálculo
x=5 # posição inicial
taxa_apren=0.9 # taxa de aprendizado
toleran=0.01 # tolerância
maxi_itera=1000 # # máximo de iterações
```

```python
# Salva as dados de trajetória
x_valor=[x]
f_valor=[tb(x)]
```

```python
# código de gradiente descendente
for i in range(maxi_itera):
    grad=der_tb(x)
    novo_x=x-grad*taxa_apren

    x_valor.append(novo_x)
    f_valor.append(tb(novo_x))

    if abs(novo_x-x) < toleran:
        print(f"convergeu a iteração {i+1}")
        break

    x=novo_x
    

```

```python
# Resultados
minim_x=x
minim_fx=tb(minim_x)

print(f"\n Valor mínimo aproximado: x={minim_x}")
print(f"Valor mínimo da função: fx={minim_fx}")

```

```python
# gerando pontos da função
x_curva=np.linspace(-6,6,400)
y_curva=tb(x_curva)

# plotando
plt.figure(figsize=(8,6))
plt.plot(x_curva,y_curva,label='f(x)=x^2-1',color='blue')
plt.plot(x_valor, f_valor,'ro--',label='Trajetótia de gradiente')
plt.scatter(minim_x,minim_fx,color='green',s=100,label='Mínimo obtido')

plt.title('Gradiente descendente de f(x)=x^2-1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
#plt.savefig('alfama.png', dpi=300)
plt.show()
```
