```python
import numpy as np
import matplotlib.pyplot as plt

# Fun√ß√£o U(x, y)
def U(x, y):
    return np.sin(x) * np.sin(y) + (x * y)**2 / 500

# Gradiente de U
def grad_U(x, y):
    dU_dx = np.cos(x) * np.sin(y) + (2 * x * y**2) / 500
    dU_dy = np.sin(x) * np.cos(y) + (2 * y * x**2) / 500
    return np.array([dU_dx, dU_dy])

# Par√¢metros do gradiente descendente
taxa = 0.1           # taxa de aprendizado
tol = 1e-5            # toler√¢ncia
max_pasos = 1000      # n√∫mero m√°ximo de passos
pos = np.array([-1.0, 1.0])  # posi√ß√£o inicial

# Armazenar trajet√≥ria e valores de U
trajectory = [pos.copy()]
U_values = [U(*pos)]

# Gradiente descendente
for _ in range(max_pasos):
    grad = grad_U(*pos)
    if np.linalg.norm(grad) < tol:
        break
    pos = pos - taxa * grad
    trajectory.append(pos.copy())
    U_values.append(U(*pos))

trajectory = np.array(trajectory)

# -------------------------
# üìà Gr√°fico da fun√ß√£o U(x_n, y_n) ao longo das epochs
plt.figure(figsize=(8, 4))
plt.plot(U_values, 'b.-')
plt.title('Converg√™ncia de U(x, y) ao longo das √©pocas')
plt.xlabel('Epoch (n)')
plt.ylabel('U(x_n, y_n)')
plt.grid(True)
plt.tight_layout()
plt.savefig('conv.png', dpi=300)
plt.show()
```
