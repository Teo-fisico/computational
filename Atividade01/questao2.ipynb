```python
import numpy as np
import matplotlib.pyplot as plt

# defina a função e sua derivada
def bs(x):
    return x**4-x**2

def deri_bs(x):
    return 4*x**3-2*x
```

```python
# Parâmetros de gradiente descendente
x=-1.3 # Ponto de partida
tax_alfa=0.3 # taxa de aprendizado
num_iter=100 # número de iterações
toleran=0.01 # tolerancia
x_valor=[x] # salva a trajetória
g_valor=[bs(x)]
```

```python
for i in range(num_iter):
    grad=deri_bs(x)
    x_novo=x-tax_alfa*grad

    x_valor.append(x_novo)
    g_valor.append(bs(x_novo))

    if abs(x_novo-x)<toleran:
        print(f"Converge a iteração {i+1}")
        break
    
    x=x_novo
 
  
```

```python
min_x=x
min_gx=bs(min_x)

#print(f"\n O valor mínimo aprox. é:x={min_x}")
print(f"\n O valor mínimo de função g({min_x})={min_gx}")
```

```python
# Plotar a função e gradiente descendente
x_val=np.linspace(-1.5,1.5,300)
y_val=bs(x_val)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x_val,y_val,label='g(x)=x.x(x-1)(x+1)',color='orange')
plt.plot(x_valor,g_valor,'-->',label='Trajetória de gradiente')
plt.scatter(min_x,min_gx,color='green',s=100,label='Mínimo encontrado')

plt.title('Gradiente descendente g(x)=x.x(x-1)(x+1)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.grid(True)
plt.savefig('q21d.png')
plt.show()
```
