```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
hbar = 1.055e-34  # constante de Planck sobre 2pi(J.s)
m = 9.109e-31     # massa de elêtron (kg)
a = 1e-10         # constante de rede (m)

# rede recíproca 
G = 2*np.pi/a

# Primeira zona de Brillouin
k = np.linspace(-G, G, 500)

```

```python
# Energia dispersão para diferentes bandas (k vetor de rede reciproca)
bands = []
num_bands = 5
for n in range(-num_bands//2, num_bands//2 + 1):
    k_des = k + n * G
    E = (hbar**2 * k_des**2) / (2 * m)
    bands.append(E)
```

```python
# gráfico
plt.figure(figsize=(8,5))
for E in bands:
    plt.plot(k / G, E / 1.6e-19, 'b')  # muda para eV
plt.xlabel(r'Vetor de onda $k / G$')
plt.ylabel('Energia (eV)')
plt.title('Estrutura de banda de um elétron livre (1D)')
plt.grid(True)
plt.axvline(x=-0.5, color='gray', linestyle='--')
plt.axvline(x=0.5, color='gray', linestyle='--')
plt.xlim(-1, 1)
plt.tight_layout()
plt.show()
```

```python
# Espaço k 2D
kx = np.linspace(-2*G, 2*G, 300)
ky = np.linspace(-2*G, 2*G, 300)
KX, KY = np.meshgrid(kx, ky)

# Energia
E2D=(hbar**2*(KX**2 + KY**2))/(2 * m)

# Gráfico
plt.figure(figsize=(6,5))
cp = plt.contourf(KX/G, KY/G, E2D / 1.6e-19, levels=50, cmap='viridis')
plt.colorbar(cp, label='Energia (eV)')
plt.xlabel(r'$k_x / G$')
plt.ylabel(r'$k_y / G$')
plt.title('Modelo de elétron livre 2D: Energia de Contorno')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
